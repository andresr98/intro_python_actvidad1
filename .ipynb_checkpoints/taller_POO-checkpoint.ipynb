### Codigo 1 ###

Se tienen un objetos de la realidad, el punto tal y como se muestra a continuación:

![punto](punto.jpg)

El siguiente diagrama UML modela el **punto** como una clase llamada ```Point```:

![uml_punto1](uml_punto1.jpg)

* El constructor de la clase inicializa las coordenadas de esta (x,y) en el valor asignado por el usuario si se le pasan argumentos o en 0 si no se le pasan argumentos
* Para esta clase se definieron los siguientes métodos:
  * **distance**: Calcula la distancia de un punto a otro.
  * **set_location**: Asigna una posición determinada al punto.
  * **distance_from_origin**: Calcula la distancia del punto al origen de coordenadas.
  * **print**: Imprime las coordenadas siguiendo el siguiente formato: (x, y)
  
Dado el diagrama UML anterior hacer lo siguiente:
1. Codificar la clase en python.
2. Definir tres objetos tipo ```Point```: ```A(0,0)```,  ```B(-3,4)```, ```C(5,8)```.
3. Imprimir los tres puntos anteriormente creados.
4. Cambiar las coordenadas de A de tal modo que la componentes x de A sea la componente x de C y la componente y de A sea la componente y de B.
5. Imprimir nuevamente los puntos.
6. Calcular las distancias de AB, BC, AC y almacenarlas en las variables d_AB, d_BC y d_AC respectivamente. Luego, imprimir estos valores.
7. Calcule la distancia de A hasta el origen e imprimala.


````python
"""
Punto 1: Definicion de la clase punto
"""

from math import sqrt

# Clase Point
class Point:  
    
    # Constructor  
    def __init__(self, x = 0, y = 0):
        self.x = x
        self.y = y

    # Metodos  
    def set_location(self, x, y):
        self.x = x    
        self.y = y  
        
    def distance_from_origin(self):
        return sqrt(self.x * self.x + self.y * self.y)
    
    def distance(self, other):    
        dx = self.x - other.x    
        dy = self.y - other.y    
        return sqrt(dx**2 + dy**2)
    
    def print(self):
        print("(" + str(self.x) + ", " + str(self.y) + ")")
        

"""
Punto 2: A(0,0), B(-3,4), C(5,8)
"""
A = Point()
B = Point(-3,4)
C = Point(5,8)

"""
Punto 3: Impresion de los puntos A, B, C
"""
print("Coordenadas de los puntos A, B y C")
print("A:")
A.print()
print("B:")
B.print()
print("C:")
C.print()

"""
Punto 4: A.x = C.x y A.y = B.y
"""
A.set_location(C.x,B.y)

"""
Punto 5: Impresion de los puntos A, B, C
"""
print()
print("Nuevas coordenadas de los puntos A, B y C")
print("A:")
A.print()
print("B:")
B.print()
print("C:")
C.print()

"""
Punto 6: Calculo e impresión de las distancias
"""
d_AB = A.distance(B)
d_BC = B.distance(C)
d_AC = A.distance(C)
print()
print("Distancia entre A y B: " + str(d_AB))
print("Distancia entre B y C:",d_BC)
print("Distancia entre C y A: %s" %(d_AC))

"""
Punto 7: Calculo de la distancia de A hasta el origen
"""
print()
print("Distancia entre A y el origen es: " + str(A.distance_from_origin()))
````

````python
"""
Punto 1: Definicion de la clase punto
"""

from math import sqrt

# Clase Point
class Point:  
    
    # Constructor  
    def __init__(self, x = 0, y = 0):
        self.x = x
        self.y = y

    # Metodos  
    def set_location(self, x, y):
        self.x = x    
        self.y = y  
        
    def distance_from_origin(self):
        return sqrt(self.x * self.x + self.y * self.y)
    
    def distance(self, other):    
        dx = self.x - other.x    
        dy = self.y - other.y    
        return sqrt(dx**2 + dy**2)
    
    def __str__(self):
        s = "(" + str(self.x) + ", " + str(self.y) + ")"
        return s
        

"""
Punto 2: A(0,0), B(-3,4), C(5,8)
"""
A = Point()
B = Point(-3,4)
C = Point(5,8)

"""
Punto 3: Impresion de los puntos A, B, C
"""
print("Coordenadas de los puntos A, B y C")
print("A:", A)
print("B: " + str(B))
print("C: %s" %(C))


"""
Punto 4: A.x = C.x y A.y = B.y
"""
A.set_location(C.x,B.y)

"""
Punto 5: Impresion de los puntos A, B, C
"""
print()
print("Nuevas coordenadas de los puntos A, B y C")
print("A:", A)
print("B:",B)
print("C:",C)

"""
Punto 6: Calculo e impresión de las distancias
"""
d_AB = A.distance(B)
d_BC = B.distance(C)
d_AC = A.distance(C)
print()
print("Distancia entre A y B: " + str(d_AB))
print("Distancia entre B y C:",d_BC)
print("Distancia entre C y A: %s" %(d_AC))

"""
Punto 7: Calculo de la distancia de A hasta el origen
"""
print()
print("Distancia entre A y el origen es: " + str(A.distance_from_origin()))
````

1. Desarrolle una clase llamada ```Temperature``` que almacene el valor de una temperatura. La clase debe tener el siguiente constructor:

```python
def __init__(self, temperature = 75)
```

La clase debe contener así mismo los métodos descritos a continuación:
* **__str__**: Debería retornar un string de la forma ```"75 degrees Farenheit"```
* **aboveFreezing()**: Retorna True si la temperatura está arriba del punto de congelación (0°C,  32°F o 273.15°K).
* **convertToFahren()**: Retorna un nuevo objeto tipo temperatura convertido a grados Fahrenheit.
* **convertToCelcius()**: Retorna un nuevo objeto tipo temperatura convertido a grados Celcius.
* **convertToKelvin()**: Retorna un nuevo objeto tipo temperatura convertido a grados Kelvin.

Como estos métodos no se implementan (pues ```Temperature``` será la clase base) coloque la función ```pass``` en el cuerpo del método, así:

```python
class XXX:
    def metodo(self):
        # Método no implementado
        pass
```

Desarrolle las subclases ```Fahrenheit```, ```Celcius``` y ```Kelvin``` para implementar apropiadamente cada uno de los métodos de la clase ```Temperature```. (Note que cuando un método de conversión que es aplicado no tiene sentido, por ejemplo, temp1.convToFarenheit() donde temp1 es un objeto tipo Fahrenheit, entonces una copia del objeto temperatura deberá ser retornada).

La siguiente tabla le puede ser de utilidad para implementar las fórmulas de conversión para cada caso:

||Grados Kelvin (K)|Grados Celcius (C)|Grados Fahrenheit (F)|
|:---|:---|:---|:---|
|**Grados Kelvin (K)**|$K = K$|$K = C + 273.15$|$K = \frac{5}{9}\left(F + 459.67 \right)$|
|**Grados Celcius (K)**|$C = k - 273.15$|$C = C$|$K = \frac{5}{9}\left(F + 32 \right)$|
|**Grados Fahrenheit (K)**|$F = \frac{9}{5}F - 459.67 $|$F = \frac{9}{5}C -32 $|$ F = F$|

Demuestre el funcionamiento de las clases realizando las siguientes tareas:
* Crear una lista de dos objetos tipo Temperature.
* Imprima la salida de cada temperatura en la lista y agregue "above freezing" si la temperatura está arriba del punto de congelación (para la respectiva escala de temperatura).
* Crear una nueva lista de temperaturas conteniendo la temperatura de la lista original a una escala de temperatura común (Farenheit, Celcius o Kelvin). 
* Para cada temperatura en la nueva lista, imprima el valor de la temperatura y si este se encuentra arriba del punto de congelación. 
  


````python
'''Se crea la clase base temperatura, que funcionará como una interfaz'''
class Temperature:
    
    def __init__(self, temperature = 75):
        self.temperature = temperature
        
    def __str__(self):
        return str(self.temperature) + " degrees"

    def aboveFreezing(self):
        pass
    
    def convertToFahren(self):
        pass
    
    def convertToCelcius(self):
        pass
    
    def convertToKelvin(self):
        pass
````

````python
'''Se crean las distintas subclases'''

class Fahrenheit(Temperature):

    def __init__(self, temperature):
        super().__init__(temperature)

    def __str__(self):
        return super().__str__() + " in Fahrenheit"
    
    def aboveFreezing(self):
        return self.temperature >= 32

    def convertToFahren(self):
        return Fahrenheit(self.temperature)

    def convertToCelcius(self):
        return Celcius((5/9) * (self.temperature + 32))

    def convertToKelvin(self):
        return Kelvin((5/9) * (self.temperature + 459.67))

    
class Celcius(Temperature):

    def __init__(self, temperature):
        super().__init__(temperature)

    def __str__(self):
        return super().__str__() + " in Celcius"
  
    def aboveFreezing(self):
        return self.temperature >= 0
    
    def convertToFahren(self):
        return Fahrenheit(((9/5)*self.temperature) - 32)

    def convertToCelcius(self):
        return Celcius(self.temperature)

    def convertToKelvin(self):
        return Kelvin(self.temperature + 273.15)

#Clase K
class Kelvin(Temperature):
  
    def __init__(self, temperature):
        super().__init__(temperature)

    def __str__(self):
        return super().__str__() + " in Kelvin"
  
    def aboveFreezing(self):
        return self.temperature >= 273.15

    def convertToFahren(self):
        return Fahrenheit(((9/5) * self.temperature) - 459.67)

    def convertToCelcius(self):
        return Celcius(self.temperature - 273.15)

    def convertToKelvin(self):
        return Kelvin(self.temperature)
````

````python
'''Listar dos temperaturas'''

lista = [Celcius(0), Fahrenheit(32)]

for temp in lista:
    if temp.aboveFreezing():
        print(str(temp) + " and is above freezing")
    else:
        print(str(temp))
````

````python
'''Lista de temperaturas para cambiarlas a una escala común'''

nuevas_temp = lista + [Kelvin(300), Celcius(-25), Fahrenheit(36), Kelvin(200), Celcius(10), Fahrenheit(-35), Kelvin(375)]
type_temps = ['Celcius', 'Fahrenheit', 'Kelvin']

def print_above_freezing(temps):
    for temp in temps:
        if temp.aboveFreezing():
            print(str(temp) + " and is above freezing")
        else:
            print(str(temp))

for type_temp in type_temps:
    print("Se pasaran todas las temperaturas a " + type_temp + '\n')
    temps = []
    
    if type_temp == 'Celcius':
        for temp in nuevas_temp:
            new_temp = temp.convertToCelcius()
            temps.append(new_temp)
    elif type_temp == "Fahrenheit":
        for temp in nuevas_temp:
            new_temp = temp.convertToFahren()
            temps.append(new_temp)
    else:
        for temp in nuevas_temp:
            new_temp = temp.convertToKelvin()
            temps.append(new_temp)

    print_above_freezing(temps)
    print('----------------------------\n')
````

2. Escriba una clase llamada Fraccion para representar un numero fraccionario (como 1/2 o -3/8).  Tenga en cuenta lo siguiente:
  * La fracción siempre debe ser almacenada en forma reducida. Por ejemplo almacene 4/12 como 1/3 o -6/9 como -2/3. No olvide que esto debe ser realizado en el constructor.
  * Sobreescriba los métodos __add__, __sub__, __mul__ ,__truediv__ y __str__ (Recuerde el [enlace](https://www.python-course.eu/python3_magic_methods.php) como fuente de consulta). Estos métodos aceptaran otra fracción como parámetro y modificaran la fracción existente operar ese parámetro.
* Haga varios casos de test para probar el correcto funcionamiento de la clase.

* **Nota**: Antes de codificar la clase en Python se debe plantear el diagrama UML asociado a la clase.

````python
'''Clase Fraccion'''

class Fraccion:
    
    def __init__(self, numerador, denominador):
        if denominador == 0:
            denominador = 1
        self.numerador, self.denominador = Fraccion.reduce(numerador, denominador)
    
    @staticmethod
    def max_comun_div(a,b):
        while b != 0:
            a, b = b, a%b
        return a
    
    @classmethod
    def reduce(cls, n1, n2):
        comun = cls.max_comun_div(n1, n2)
        return (n1 // comun, n2 // comun)
    
    def __str__(self):
        return str(self.numerador)+'/'+str(self.denominador)
    
    '''Overriding methods'''
    
    def __add__(self, other):
        numerador = (self.numerador * other.denominador) + (self.denominador * other.numerador)
        denominador = self.denominador * other.denominador
        return Fraccion(numerador, denominador)
    
    def __sub__(self, other):
        numerador = (self.numerador * other.denominador) - (self.denominador * other.numerador)
        denominador = self.denominador * other.denominador
        return Fraccion(numerador, denominador)
    
    def __mul__(self, other):
        numerador = self.numerador * other.numerador
        denominador = self.denominador * other.denominador
        return Fraccion(numerador, denominador)
    
    def __truediv__(self, other):
        numerador = self.numerador * other.denominador
        denominador = self.denominador * other.numerador
        return Fraccion(numerador, denominador)
````

````python
'''Se probaran las 4 operaciones sobrecargadas'''

x = Fraccion(8, 5)
y = Fraccion(29, 7)

print(x)
print(y)

suma = x + y
print("La suma es: " + str(suma))

resta = x - y
print("La resta es: " + str(resta))

multiplicacion = x * y
print("La multiplicacion es: " + str(multiplicacion))

division = x / y
print("La división es: " + str(division))
````
